```python
import pandas as pd

data = {
    'Name': ['John', 'Mary', 'David', 'Sam'],
    'Sales': [1000, None, 1500, None]
}

df = pd.DataFrame(data)

print(df)
```

MISSING VALUES NaN

```python
print(df.isnull())
```

dropna

```python
clean_df = df.dropna()

print(clean_df)
```

fillna

```python
df.fillna(0, inplace=True)

print(df)
```

```python
data = {
    'Name': ['John', 'Mary', 'David', 'Sam'],
    'Marks': [50, 60, 70, None]
}


df = pd.DataFrame(data)

mean_marks = df['Marks'].mean()

df['Marks'] = df['Marks'].fillna(mean_marks)

print(df)
```

```python
import pandas as pd

data = {
    'Region': ['South', 'North', 'East', 'West', 'South', 'North'],
    'Products': ['Laptop', 'Mobile', 'Laptop', 'Tablet', 'Mobile', 'Tablet'],
    'Sales': [50000, 30000, 45000, 30000, 20000, 35000]
}

df = pd.DataFrame(data)

print(df)

south_sales = df[df["Region"] == "South"]

print(south_sales)
```

```python
sorted_df = df.sort_values("Sales")
print(sorted_df)
```

```python
south_sales = df[df["Region"] == "South"]
print(south_sales)
```

```python
high_sales = df[df["Sales"] > 30000]
print(high_sales)
```

```python
import matplotlib.pyplot as plt

months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun"]
sales = [10000, 12000, 15000, 13000, 18000, 20000]

plt.figure(figsize=(12,5))  # Makes the whole figure wider

# Line Chart - Left side
plt.subplot(1,2,1)  # 1 row, 2 columns, position 1
plt.plot(months, sales, marker="o")
plt.title("Sales Trend")
plt.xlabel("Months")
plt.ylabel("Sales")

# Bar Chart - Right side
plt.subplot(1,2,2)  # 1 row, 2 columns, position 2
plt.bar(months, sales, color="orange")
```

```python
import matplotlib.pyplot as plt

months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun"]
sales = [10000, 12000, 15000, 13000, 18000, 20000]

plt.figure(figsize=(12,5))

# Line Chart
plt.subplot(1,2,1)
plt.plot(months, sales, marker="o", color="#1f77b4", linewidth=2)
plt.title("Sales Trend")
plt.xlabel("Months")
plt.ylabel("Sales")
plt.grid(True, linestyle='--', alpha=0.6)

# Bar Chart
plt.subplot(1,2,2)
plt.bar(months, sales, color="orange", alpha=0.8)
plt.title("Monthly Sales")
plt.xlabel("Months")
plt.ylabel("Sales")

plt.suptitle("Sales Performance Jan-Jun 2026", fontsize=16, y=1.02)
plt.tight_layout()  # Fixes spacing
plt.show()
```

```python
import matplotlib.pyplot as plt

months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun"]
sales = [10000, 12000, 15000, 13000, 18000, 20000]

# Line Chart
plt.subplot(1, 2, 1)
plt.plot(months, sales, marker="o")
plt.title("Sales Trend")
plt.xlabel("Months")
plt.ylabel("Sales")

# Bar Chart
plt.subplot(1, 2, 2)
plt.bar(months, sales, color="orange")
plt.title("Sales Comparison")
plt.xlabel("Months")
plt.ylabel("Sales")

plt.show()

# Pie Chart
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun"]
sales = [10000, 12000, 15000, 13000, 18000, 20000]

plt.pie(sales, labels=months, autopct="%1.1f%%")
plt.title("Sales Distribution")

plt.show()
```
