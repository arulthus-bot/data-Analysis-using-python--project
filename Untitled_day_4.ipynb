```python
import numpy as np

marks = np.array([70,80,90,60])

average = np.mean(marks)

print(average)
```

```python
salary = np.array([10000,20000,30000,40000,50000])
med = np.median(salary)
print(med)
```

```python
# 1. Define the sales data array
sales = [12000, 15000, 18000, 14000, 20000]

# 2. Perform calculations
total_sales = sum(sales)
average_sales = total_sales / len(sales)
highest_sales = max(sales)
lowest_sales = min(sales)

# 3. Print the results
print("Monthly Sales:", sales)
print("Total Sales:", total_sales)
print("Average Sales:", average_sales)
print("Highest Sales:", highest_sales)
print("Lowest Sales:", lowest_sales)
```

```python
scores = [80, 90, 78, 95, 88]

total = sum(scores)
average = total / len(scores)
highest = max(scores)
lowest = min(scores)

print("Total Score:", total)
print("Average Score:", average)
print("Highest Score:", highest)
print("Lowest Score:", lowest)
```
